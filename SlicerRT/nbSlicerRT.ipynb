# SLICER RT

Slicer propose des extensions, notamment une extension RT.

- Télécharger le module Slicer RT

## Présentation

Pour présenter cette extension, on se propose d'évaluer des contours et des matrices de dose. 

### Contexte clinique de l'exemple

- Traitement du cancer de la prostate : Protocole RPAH2
- 62Gy/20s vs 37,5Gy/5s
- Injection d’acide hyaluronique entre la face postérieure de la prostate et la face antérieure du rectum
- Implants intra-prostatiques
- CBCT quotidien
- Recalage quotidien sur les implants

### Exemple pour un patient du bras à 5 séances

Le module Slicer RT permet de visualiser facilement dans un même référentiel l’ensemble des images et des contours (recalage, fusion) et de comparer des contours ou des matrices de doses.

### Exercices
Les exercices suivants permettent de prendre en main les fonctionnalités de recalage et de comparaison.


## Effectuer un recalage rigide

### Recalage
- Load DCM, PAT1, CBCT1 + RTStruct et CBCT3 + RTStruct
- Choisir le module Data
- Dans Display & Modify Scene>Nodes : Cliquer sur 1: Unknown, Cliquer droit : Rename, taper : CBCT1
- Dans Display & Modify Scene>Nodes : Cliquer sur 1: Unknown_1, Cliquer droit : Rename, taper : CBCT3

- Choisir le module : Registration>Transforms
- Active Transform: Cliquer Select a LinearTransform
- Active Transform: Choisir Create and rename new LinearTransform
- Active Transform: CBCT3_CBCT1
- Apply transform : Surligner CBCT3 et les contours associés et cliquer la sur flèche verte vers la droite
- Edit, Translation : Ajuster les valeurs LR, PA et IS pour effectuer un recalage en translation.

### Comparaison de contour
Après avoir réalisé un recalage, il est possible de comparer des contours par exemple en utilisant l’indice DICE.

- Choisir le module Radiotherapy>Contour Comparison
- Choisir les contours à comparer sur une imagerie de référence et calculer le coefficient DICE (DSC=2xA∩B/(A+B)).
- Calculer le DSC sur le contour prostate après recalage osseux
- Calculer le DSC sur le contour prostate après recalage sur implants

### Comparaison des matrices de dose
Il est également possible de comparer des distributions de dose (HDV, isodoses, différence, gamma).

- Pour l’exemple, charger les matrices de dose associées aux CBCT1 et CBCT3.
- Ajouter la matrice de dose CBCT3 dans les objets liés à la transformation.
- Dans le module Radiotherapy>Dose Volume Histogram, choisir la matrice de dose du CBCT1 et le set de contour. Calculer.
- Recommencer avec le CBCT3.


## Créer un masque à partir d’un contour

- Load DCM, PAT1, CBCT1 + RTStruct
- Choisir le module : Radiotherapy>Contours
- Current contour : Choisir ProstateCBCT1_Contour
- Change active representation > Convert to: Indexed labelmap
- Reference volume 1:Unknown
- Cliquer Apply
- Contour functions : Extract labelmap from contour : Cliquer Extract
- La carte ProstateCBCT1_Contour_Indexedlabelmap a été créée. Il est possible de la visualiser seule (B) ou en superposition avec le CBCT1 (L). On peut s’apercevoir qu’il y a des petits défauts. Donc on reprend la technique (dilatation, érosion)

- Choisir le module : Filtering>Simple Filters
- Dans Filters : Choisir GrayscaleDilateImageFilter
- Dans Parameters : Input Volume : ProstateCBCT1_Contour_Indexedlabelmap
- Dans Parameters : Output Volume : Create and rename new Volume : ProstateCBCT1_Dilate
- Cocher Labelmap
- Cliquer Apply
- Choisir le module : Filtering>Simple Filters
- Dans Filters : Choisir GrayscaleErodeImageFilter
- Dans Parameters : Input Volume : ProstateCBCT1_Dilate
- Dans Parameters : Output Volume : Create and rename new Volume : ProstateCBCT1_Map
- Cocher Labelmap
- Cliquer Apply

- La carte ProstateCBCT1_Map correspond à present au contour.

### Probabilité de présence
- Si cette opération est répétée sur plusieurs CBCT, il est possible de créer le masque de la prostate pour toutes les séances, d’additionner les masques et de comparer à l’isodose 95% ou au PTV… Pour réaliser cela, il faut également connaitre comment additionner des images.
- Pour additionner deux images : Filtering>Arithmetic>Add Scalar Volumes
- L'addition des masques donnera le nombre de présence d'un voxel sur tout le traitement.
